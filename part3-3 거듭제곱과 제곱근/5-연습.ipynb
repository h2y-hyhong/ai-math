```python
# 다음 수식을 그래프로 그려보자
from IPython.display import display, Math

display(Math(r'y = \sqrt{x} + 1'))

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x) + 1

x = np.linspace(0, 9)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
