```python
# 제곱근의 그래프를 그린다
from IPython.display import display, Math

display(Math(r'y = \sqrt{x}'))

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x)

x = np.linspace(0, 9)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
