```python
# 거듭제곱의 그래프를 그린다
from IPython.display import display, Math

display(Math(r'y = x^2'))

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


def my_func(x):
    a = 3
    return x ** a

x = np.linspace(0, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
