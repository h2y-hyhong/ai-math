```python
# 코드를 보완하고 다음 수식의 그래프를 그려보자
from IPython.display import display, Math

display(Math(r'y = x ^ 3 - 2x ^ 2 - 3x + 3'))

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return x**3 - 2*x** 2 - 3*x + 3

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
