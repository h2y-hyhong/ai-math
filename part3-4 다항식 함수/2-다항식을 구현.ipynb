```python
# 2차 다항식을 그래프로 그린다
from IPython.display import display, Math

display(Math(r'y = 3x^2 + 2x + 1'))

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


def my_func(x):
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
# 3차 다항식을 그래프로 그린다
from IPython.display import display, Math

display(Math(r'y = 4x ^ 3 + 2x ^ 2 + x + 3'))

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return 4*x**3 + 2*x**2 + x + 3

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
