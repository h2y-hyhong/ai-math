```python
# 문제

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3)
y_1 = 1.5 * x
y_2 = -2 * x + 1

plt.xlabel("x value", size = 14)
plt.ylabel("y value", size = 14)

plt.title("My Graph")

plt.grid()

plt.plot(x, y_1, label = "y1")
plt.plot(x, y_2, label = "y2", linestyle = "dashed")
plt.legend()

plt.show()
```
